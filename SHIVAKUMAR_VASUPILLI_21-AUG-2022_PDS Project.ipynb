### In E-commerce, both the retailers (here brands) and the company have to make profit to sustain in the business. The E-Commerce company has the following rule for computing their own revenue:
    The company charges
    (i) 25% on the orders having final price (discounted price) greater than 600
    (ii) 15% on the orders having final price (discounted price) greater than 350 but less than or equal to 600
    (iii) 10% on the orders having final price (discounted price) greater than 100 but less than or equal to 350
    (iv) Otherwise, 5% on the final price (discounted price) 

### Q 5. Find the net revenue generated by the E-Commerce company over all orders placed.

```python
# Creating a if condition to find the E_commerce revenue
ECom_Data['E_Commerce_Revenue']= [i*0.25 if i>600 else i*0.15 if 350<i<=600 else i*0.1 if 100<i<=350 else i*0.05 for i in ECom_Data.discounted_price]
ECom_Data.filter(['retail_price', 'discounted_price', 'E_Commerce_Revenue']).head(20)
```

```python
# Net revenue generated by the Ecommerce company
Net_revenue=ECom_Data['E_Commerce_Revenue'].sum()
Net_revenue
```

### Ans 5) Explanation:The Net Revenue of the E-Commerce company can be calculated by applying the "if condition"  and finding the revenue generated from each product by applying the companies condition and then adding all those values. 
### Result: The Net revenue generated by the E-Commerce company is rupees 22,42,221.75

### From the above data we can say that the date is in object data type but it needs to be in date time function. 

```python
# changing the data type of Order_date 
ECom_Data['Order_Date']=pd.to_datetime(ECom_Data['Order_Date'], dayfirst=True)
```

```python
ECom_Data.info()
```

### Therefore the Order_Date datatype was changed to "datetime"

```python
#extracting ordered_month and creating a new ordered_month column
ECom_Data['ordered_month']=ECom_Data['Order_Date'].dt.month
ECom_Data
```

### The above graph shows the lineplot for the monthly net Revenue of E-Commerce Company for each region
